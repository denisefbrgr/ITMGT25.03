```python
import requests
import time
import matplotlib.pyplot as plt

latitudes = []
longitudes = []

data_points = 0

while data_points < 100:
    print(data_points)
    response = requests.get('http://api.open-notify.org/iss-now.json')
    
    if response.status_code == 200:
        data_json = response.json()
        latitude = float(data_json['iss_position']['latitude'])
        longitude = float(data_json['iss_position']['longitude'])
        latitudes.append(latitude)
        longitudes.append(longitude)
        data_points += 1
    else:
        print("Error in retrieving ISS location data.")
        
    time.sleep(10)

plt.scatter(longitudes, latitudes, c='blue', label = 'Location')
plt.title("International Space Station Path")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```
